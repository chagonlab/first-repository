```python
from ast import Continue
import random

answer = random.sample(range(0, 10), 5) # list 안에 int
print(answer)

while True :
    user_input = input('0에서 10까지의 숫자 중 5개를 중복없이 골라주세요') # str로 인풋 받음
    users = list() # user의 input data를 list로 저장하여 answer와 비교 가능하게 만듬

    for i in user_input :
        users.append(int(i)) # 반복문을 통해 users 리스트에 int 값으로 append

    if len(users) == 5 and len(set(users)) == 5 : # 5글자인지, 중복이 없는지
        strike = 0
        ball = 0
        for i in range(len(answer)) :
          if answer[i] == users[i] :
            strike += 1
          elif users[i] in answer :
            ball += 1

    elif len(users) != 5 :
        print('5글자인지 다시 확인해주세요')
        continue

    elif len(set(users)) != 5 :
        print('중복이 있는지 확인해주세요')
        continue

    print(strike, 'strike', ball, 'ball')
    break



```
